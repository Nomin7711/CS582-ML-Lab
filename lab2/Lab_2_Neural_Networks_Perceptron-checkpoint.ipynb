## Question 1: Threshold Neuron with 2 Inputs

**Given:** A neuron with 2 inputs, 1 output, threshold activation function.  
- Weights: $w_1 = 1$, $w_2 = 1$  
- Bias: $b = -1.5$  
- Activation: threshold (step) function: $y = 1$ if $z \geq 0$, else $y = 0$, where $z = w_1 x_1 + w_2 x_2 + b$

**Compute net input and output for each input pair:**

| $x_1$ | $x_2$ | $z = w_1 x_1 + w_2 x_2 + b$ | $z$ value | Output $y$ |
|-------|-------|-------------------------------|-----------|------------|
| 0     | 0     | $1(0) + 1(0) - 1.5$          | $-1.5$    | **0**      |
| 1     | 0     | $1(1) + 1(0) - 1.5$          | $-0.5$    | **0**      |
| 0     | 1     | $1(0) + 1(1) - 1.5$          | $-0.5$    | **0**      |
| 1     | 1     | $1(1) + 1(1) - 1.5$          | $0.5$     | **1**      |

**Truth table comparison:** This matches the **AND** gate — output is 1 only when both inputs are 1.

**Discriminant function equation:**

The decision boundary is where $z = 0$:

$$w_1 x_1 + w_2 x_2 + b = 0 \implies x_1 + x_2 - 1.5 = 0 \implies x_2 = -x_1 + 1.5$$

This is a line with slope $-1$ and y-intercept $1.5$. Points above/on the line output 1; points below output 0.

```python
import numpy as np
import matplotlib.pyplot as plt

# Q1: Plot the discriminant function for the AND gate
fig, ax = plt.subplots(1, 1, figsize=(6, 6))

# Decision boundary: x1 + x2 - 1.5 = 0 => x2 = -x1 + 1.5
x_line = np.linspace(-0.5, 2.0, 100)
y_line = -x_line + 1.5

# Input points
inputs = np.array([[0,0],[1,0],[0,1],[1,1]])
outputs = np.array([0, 0, 0, 1])

ax.plot(x_line, y_line, 'k-', linewidth=2, label='Decision boundary: $x_1 + x_2 = 1.5$')
ax.scatter(inputs[outputs==0, 0], inputs[outputs==0, 1], c='red', s=150, marker='x', linewidths=2, label='Output = 0')
ax.scatter(inputs[outputs==1, 0], inputs[outputs==1, 1], c='blue', s=150, marker='o', linewidths=2, label='Output = 1')

ax.fill_between(x_line, y_line, 2.5, alpha=0.1, color='blue')
ax.fill_between(x_line, -0.5, y_line, alpha=0.1, color='red')

ax.set_xlim(-0.5, 2.0)
ax.set_ylim(-0.5, 2.0)
ax.set_xlabel('$x_1$', fontsize=14)
ax.set_ylabel('$x_2$', fontsize=14)
ax.set_title('Q1: AND Gate - Decision Boundary', fontsize=14)
ax.legend()
ax.grid(True, alpha=0.3)
ax.set_aspect('equal')
plt.tight_layout()
plt.show()
```

## Question 3: Perceptron Convergence Theorem with $\|x\| \leq R$

**Theorem (Perceptron Convergence Theorem, generalized):**  
If the training data is linearly separable, the perceptron learning algorithm will converge in a finite number of updates.

**Setup:**  
- Training set: $\{(x_i, t_i)\}$ where $t_i \in \{-1, +1\}$
- Assume $\|x_i\| \leq R$ for all $i$ (instead of $\|x\| \leq 1$)
- Since data is linearly separable, there exists a weight vector $w^*$ with $\|w^*\| = 1$ such that $t_i \cdot (w^{*T} x_i) \geq \gamma > 0$ for all $i$ (margin $\gamma$)

**Proof:**

Let $w^{(k)}$ denote the weight vector after $k$ updates. Initialize $w^{(0)} = 0$.

The perceptron update rule (on misclassified example $(x_i, t_i)$):
$$w^{(k+1)} = w^{(k)} + t_i x_i$$

**Step 1: Lower bound on $w^{*T} w^{(k)}$ (grows linearly with $k$)**

$$w^{*T} w^{(k+1)} = w^{*T} (w^{(k)} + t_i x_i) = w^{*T} w^{(k)} + t_i (w^{*T} x_i)$$

Since $t_i (w^{*T} x_i) \geq \gamma$:
$$w^{*T} w^{(k+1)} \geq w^{*T} w^{(k)} + \gamma$$

By induction (starting from $w^{*T} w^{(0)} = 0$):
$$w^{*T} w^{(k)} \geq k\gamma \quad \cdots (1)$$

**Step 2: Upper bound on $\|w^{(k)}\|^2$ (grows linearly with $k$)**

$$\|w^{(k+1)}\|^2 = \|w^{(k)} + t_i x_i\|^2 = \|w^{(k)}\|^2 + 2t_i(w^{(k)T} x_i) + \|x_i\|^2$$

Since $(x_i, t_i)$ was misclassified, $t_i(w^{(k)T} x_i) \leq 0$, so:
$$\|w^{(k+1)}\|^2 \leq \|w^{(k)}\|^2 + \|x_i\|^2 \leq \|w^{(k)}\|^2 + R^2$$

By induction (starting from $\|w^{(0)}\|^2 = 0$):
$$\|w^{(k)}\|^2 \leq kR^2 \quad \cdots (2)$$

**Step 3: Combine the bounds**

By Cauchy-Schwarz: $w^{*T} w^{(k)} \leq \|w^*\| \cdot \|w^{(k)}\| = \|w^{(k)}\|$

From (1): $k\gamma \leq \|w^{(k)}\|$

Squaring: $k^2 \gamma^2 \leq \|w^{(k)}\|^2$

From (2): $k^2 \gamma^2 \leq kR^2$

Therefore:
$$k \leq \frac{R^2}{\gamma^2}$$

**Conclusion:** The number of weight updates $k$ is bounded above by $\frac{R^2}{\gamma^2}$, so the algorithm converges in a finite number of steps. $\blacksquare$

**Note:** When $\|x\| \leq 1$ (i.e., $R = 1$), this reduces to the standard bound $k \leq \frac{1}{\gamma^2}$.

## Question 4: Batch vs Sequential Perceptron Update

The batch update algorithm accumulates all errors over the entire dataset before updating weights once per epoch. The sequential (online) update modifies weights after each individual training example.

Below we implement both versions and compare their convergence on the AND gate.

```python
import numpy as np

def perceptron_batch(X, targets, learning_rate=0.25, max_epochs=100):
    """Batch update perceptron: weights updated once per epoch using sum of all errors."""
    n_samples, n_features = X.shape
    # Add bias column
    X_bias = np.concatenate([X, -np.ones((n_samples, 1))], axis=1)
    weights = np.random.uniform(-0.5, 0.5, n_features + 1)
    
    errors_per_epoch = []
    
    for epoch in range(max_epochs):
        # Forward pass for all samples
        activations = np.where(np.dot(X_bias, weights) > 0, 1, 0)
        errors = targets - activations
        n_errors = np.sum(np.abs(errors))
        errors_per_epoch.append(n_errors)
        
        if n_errors == 0:
            break
        
        # Batch update: accumulate weight changes from ALL samples, then update once
        weight_update = learning_rate * np.dot(errors, X_bias)
        weights += weight_update
    
    return weights, errors_per_epoch


def perceptron_sequential(X, targets, learning_rate=0.25, max_epochs=100):
    """Sequential (online) update perceptron: weights updated after each sample."""
    n_samples, n_features = X.shape
    # Add bias column
    X_bias = np.concatenate([X, -np.ones((n_samples, 1))], axis=1)
    weights = np.random.uniform(-0.5, 0.5, n_features + 1)
    
    errors_per_epoch = []
    
    for epoch in range(max_epochs):
        n_errors = 0
        
        for i in range(n_samples):
            # Forward pass for single sample
            activation = 1 if np.dot(X_bias[i], weights) > 0 else 0
            error = targets[i] - activation
            
            if error != 0:
                n_errors += 1
                # Sequential update: update weights immediately after each sample
                weights += learning_rate * error * X_bias[i]
        
        errors_per_epoch.append(n_errors)
        
        if n_errors == 0:
            break
    
    return weights, errors_per_epoch


# Test on AND gate
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
targets = np.array([0, 0, 0, 1])

np.random.seed(42)
w_batch, err_batch = perceptron_batch(X, targets)
np.random.seed(42)
w_seq, err_seq = perceptron_sequential(X, targets)

print("=== AND Gate Results ===")
print(f"\nBatch Update:      weights = {w_batch}, converged in {len(err_batch)} epochs")
print(f"Sequential Update: weights = {w_seq}, converged in {len(err_seq)} epochs")

# Verify outputs
X_bias = np.concatenate([X, -np.ones((4, 1))], axis=1)
print(f"\nBatch outputs:      {np.where(np.dot(X_bias, w_batch) > 0, 1, 0)}")
print(f"Sequential outputs: {np.where(np.dot(X_bias, w_seq) > 0, 1, 0)}")
print(f"Expected:           {targets}")
```

```python
# Compare convergence across multiple logic gates
gates = {
    'AND': np.array([0, 0, 0, 1]),
    'OR':  np.array([0, 1, 1, 1]),
    'NAND': np.array([1, 1, 1, 0]),
    'NOR': np.array([1, 0, 0, 0]),
}

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

for ax, (gate_name, gate_targets) in zip(axes.flat, gates.items()):
    batch_epochs_list = []
    seq_epochs_list = []
    
    for seed in range(50):
        np.random.seed(seed)
        _, err_b = perceptron_batch(X, gate_targets)
        np.random.seed(seed)
        _, err_s = perceptron_sequential(X, gate_targets)
        batch_epochs_list.append(len(err_b))
        seq_epochs_list.append(len(err_s))
    
    ax.bar([0, 1], [np.mean(batch_epochs_list), np.mean(seq_epochs_list)],
           yerr=[np.std(batch_epochs_list), np.std(seq_epochs_list)],
           color=['steelblue', 'coral'], capsize=5, width=0.5)
    ax.set_xticks([0, 1])
    ax.set_xticklabels(['Batch', 'Sequential'])
    ax.set_ylabel('Epochs to converge')
    ax.set_title(f'{gate_name} Gate')
    ax.grid(axis='y', alpha=0.3)

fig.suptitle('Q4: Batch vs Sequential Perceptron - Average Epochs to Converge (50 seeds)', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()

print("\nAverage epochs across 50 random seeds:")
print(f"{'Gate':<6} {'Batch':>10} {'Sequential':>12}")
print("-" * 30)
for gate_name, gate_targets in gates.items():
    b_list, s_list = [], []
    for seed in range(50):
        np.random.seed(seed)
        _, err_b = perceptron_batch(X, gate_targets)
        np.random.seed(seed)
        _, err_s = perceptron_sequential(X, gate_targets)
        b_list.append(len(err_b))
        s_list.append(len(err_s))
    print(f"{gate_name:<6} {np.mean(b_list):>10.1f} {np.mean(s_list):>12.1f}")
```

### Q4 Observations

- **Sequential updates tend to converge in fewer epochs** than batch updates because the weights are corrected immediately after each misclassified sample, allowing subsequent samples in the same epoch to benefit from the correction.
- **Batch updates** are more stable since they average gradients across the full dataset, but this means corrections are delayed, often requiring more epochs.
- Both methods converge to correct weights for all linearly separable logic gates (AND, OR, NAND, NOR).
- The final weight values may differ between methods, but both produce valid decision boundaries.

## Question 5: Perceptron on the Pima Indians Diabetes Dataset

We load the Pima Indians Diabetes dataset, explore it, normalize features, split into train/test, train a perceptron, and analyze the results.

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the Pima Indians Diabetes dataset
# Columns: Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin, BMI, DiabetesPedigreeFunction, Age, Outcome
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"

try:
    data = np.genfromtxt(url, delimiter=',')
except Exception:
    # If URL fails, try local file
    data = np.genfromtxt('pima-indians-diabetes.csv', delimiter=',')

feature_names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
                 'Insulin', 'BMI', 'DiabetesPedigree', 'Age']

X = data[:, :-1]
y = data[:, -1].astype(int)

print(f"Dataset shape: {X.shape}")
print(f"Class distribution: {np.sum(y==0)} non-diabetic, {np.sum(y==1)} diabetic")
print(f"\nFeature statistics:")
print(f"{'Feature':<22} {'Mean':>8} {'Std':>8} {'Min':>8} {'Max':>8}")
print("-" * 50)
for i, name in enumerate(feature_names):
    print(f"{name:<22} {X[:,i].mean():>8.1f} {X[:,i].std():>8.1f} {X[:,i].min():>8.1f} {X[:,i].max():>8.1f}")
```

```python
# Explore the data: plot feature distributions by class
fig, axes = plt.subplots(2, 4, figsize=(16, 8))

for i, (ax, name) in enumerate(zip(axes.flat, feature_names)):
    ax.hist(X[y==0, i], bins=20, alpha=0.6, label='Non-diabetic', color='steelblue')
    ax.hist(X[y==1, i], bins=20, alpha=0.6, label='Diabetic', color='coral')
    ax.set_title(name)
    ax.legend(fontsize=8)

fig.suptitle('Q5: Feature Distributions by Class', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Normalize features and split into train/test
def normalize(X):
    """Min-max normalization to [0, 1]."""
    X_min = X.min(axis=0)
    X_max = X.max(axis=0)
    return (X - X_min) / (X_max - X_min + 1e-8)

X_norm = normalize(X)

# Shuffle and split: 80% train, 20% test
np.random.seed(42)
indices = np.random.permutation(len(X_norm))
split = int(0.8 * len(X_norm))

X_train, X_test = X_norm[indices[:split]], X_norm[indices[split:]]
y_train, y_test = y[indices[:split]], y[indices[split:]]

print(f"Train set: {X_train.shape[0]} samples")
print(f"Test set:  {X_test.shape[0]} samples")
```

```python
# Train perceptron on Pima Indians dataset (using both batch and sequential)

def perceptron_train(X, targets, mode='sequential', learning_rate=0.25, max_epochs=200):
    """Train perceptron with batch or sequential updates. Returns weights and error history."""
    n_samples, n_features = X.shape
    X_bias = np.concatenate([X, -np.ones((n_samples, 1))], axis=1)
    weights = np.random.uniform(-0.1, 0.1, n_features + 1)
    
    errors_per_epoch = []
    
    for epoch in range(max_epochs):
        if mode == 'batch':
            activations = np.where(np.dot(X_bias, weights) > 0, 1, 0)
            errors = targets - activations
            n_errors = np.sum(np.abs(errors))
            errors_per_epoch.append(n_errors)
            if n_errors == 0:
                break
            weights += learning_rate * np.dot(errors, X_bias)
        else:  # sequential
            n_errors = 0
            for i in range(n_samples):
                activation = 1 if np.dot(X_bias[i], weights) > 0 else 0
                error = targets[i] - activation
                if error != 0:
                    n_errors += 1
                    weights += learning_rate * error * X_bias[i]
            errors_per_epoch.append(n_errors)
            if n_errors == 0:
                break
    
    return weights, errors_per_epoch


def perceptron_predict(X, weights):
    """Predict using trained perceptron."""
    X_bias = np.concatenate([X, -np.ones((X.shape[0], 1))], axis=1)
    return np.where(np.dot(X_bias, weights) > 0, 1, 0)


# Train both versions
np.random.seed(42)
w_batch, err_batch = perceptron_train(X_train, y_train, mode='batch', max_epochs=200)
np.random.seed(42)
w_seq, err_seq = perceptron_train(X_train, y_train, mode='sequential', max_epochs=200)

# Evaluate on test set
pred_batch = perceptron_predict(X_test, w_batch)
pred_seq = perceptron_predict(X_test, w_seq)

acc_batch = np.mean(pred_batch == y_test) * 100
acc_seq = np.mean(pred_seq == y_test) * 100

print("=== Pima Indians Diabetes Dataset Results ===")
print(f"\nBatch Perceptron:")
print(f"  Epochs run: {len(err_batch)}")
print(f"  Final training errors: {err_batch[-1]}/{X_train.shape[0]}")
print(f"  Test accuracy: {acc_batch:.1f}%")

print(f"\nSequential Perceptron:")
print(f"  Epochs run: {len(err_seq)}")
print(f"  Final training errors: {err_seq[-1]}/{X_train.shape[0]}")
print(f"  Test accuracy: {acc_seq:.1f}%")
```

```python
# Plot training error convergence and confusion matrices
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Error convergence
axes[0].plot(err_batch, label='Batch', color='steelblue', linewidth=1.5)
axes[0].plot(err_seq, label='Sequential', color='coral', linewidth=1.5)
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Number of misclassified samples')
axes[0].set_title('Training Error Convergence')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Confusion matrix helper
def plot_confusion(ax, y_true, y_pred, title):
    tp = np.sum((y_true == 1) & (y_pred == 1))
    tn = np.sum((y_true == 0) & (y_pred == 0))
    fp = np.sum((y_true == 0) & (y_pred == 1))
    fn = np.sum((y_true == 1) & (y_pred == 0))
    cm = np.array([[tn, fp], [fn, tp]])
    
    ax.imshow(cm, cmap='Blues', interpolation='nearest')
    for i in range(2):
        for j in range(2):
            ax.text(j, i, str(cm[i, j]), ha='center', va='center', fontsize=18, fontweight='bold')
    ax.set_xticks([0, 1])
    ax.set_yticks([0, 1])
    ax.set_xticklabels(['Non-diabetic', 'Diabetic'])
    ax.set_yticklabels(['Non-diabetic', 'Diabetic'])
    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')
    ax.set_title(title)

plot_confusion(axes[1], y_test, pred_batch, f'Batch (Acc: {acc_batch:.1f}%)')
plot_confusion(axes[2], y_test, pred_seq, f'Sequential (Acc: {acc_seq:.1f}%)')

plt.tight_layout()
plt.show()
```

### Q5 Observations

1. **The perceptron does not fully converge** on this dataset — the training error plateaus but never reaches zero. This is expected because the Pima Indians Diabetes dataset is **not linearly separable** (the two classes overlap significantly in feature space).

2. **Test accuracy is around 65-75%**, which is reasonable for a simple linear classifier on this dataset. More sophisticated methods (logistic regression, MLP, SVM) typically achieve 75-80%.

3. **The training error curve oscillates** rather than monotonically decreasing. This is the **Perceptron Cycling Theorem** in action — when data is not linearly separable, the perceptron will cycle through weight states without converging.

4. **Feature distributions show overlap** between diabetic and non-diabetic classes across all features, confirming that no single linear boundary can perfectly separate the classes. Glucose and BMI show the most separation between classes.

5. **Sequential vs batch** may show different final accuracies since neither converges — the final weights depend on when training stops and the update order.

6. **Normalization is critical** — without it, features with large ranges (e.g., Insulin: 0-846) would dominate the weight updates over features with small ranges (e.g., DiabetesPedigree: 0.08-2.42).
